## Main Results of the paper

The paper covers the recent advances in Graph Transformers (GT) and compares it to Message Passing Neural Networks (MPNN). GT's have shown to outperform MPNN's and approximate them. As GNN's struggle with oversmoothing (e.g deep Convolutional layers will produce same features at different nodes after too many layers) the paper focuses on reducing this issue by introducing a Virtual Node (VN).
This VN behaves like an arbitrary node in the graph except that it is connected to all other nodes and thereby acts like a gloabl information hub.
On the theoretical side the paper proofs that this extension is able to simulate certain forms of Graph transformers.
The experimental side also suggests that this might be a good way to improve overall model performance.

The following graph shows the first experiment of the paper. Covering the performance gains by adding a VN for graph classification tasks (peptides-func) and graph regression (peptides-struct). This shows the overall performance in long range interaction modeling tasks.

```python
from experiment1.results_processing import *

results_folder = "experiment1/results_pre_trained/"

json_results = get_stats_json_paths(results_folder)
print("model count: " + str(len(json_results)))

#  To be fair for all models we only consider the best result within the same training time among all models.
min_train_time = find_min_train_time(results_folder)
print("min train time in min: " + str(min_train_time/60.0))

# Now find the best stat for each model.
best_stats = find_best_stat(json_results, min_train_time)

# generate tables:
table_func = generate_table("peptides-func", best_stats)
table_struct = generate_table("peptides-struct", best_stats)
```

#### create the plot:

```python
# Create a figure and an axis

fig, ax = plt.subplots(1, 2, figsize=(12, 2.5)) 

def plot_table(pos, data, name):
    cellText = data.values.tolist()  
    colLabels = data.columns.tolist()
    
    ax[pos].table(cellText=cellText, colLabels=colLabels, loc='center', colWidths=[0.3, 0.3, 0.4])
    ax[pos].set_title(name, pad=2)
    ax[pos].axis('off')

plot_table(0, table_func, "pepdtides-func")
plot_table(1, table_struct, "pepdtides-struct")
plt.tight_layout()
plt.show()
```

When comparing to the table from the paper it gets clear that the results are very similar. My models seem to have performed better overall but the difference between models with and without virtual nodes is clearly visible. Similar to the results of the paper all models were improved by a virtual node.
